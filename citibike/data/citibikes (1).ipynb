```python
from zipfile import ZipFile
import requests
import io

# Here we are getting the zip file
citibikeZipFile = requests.get('https://s3.amazonaws.com/tripdata/202109-citibike-tripdata.csv.zip')

# Here we are uncompressing it and reading the whole package into a variable
zipFiles = ZipFile(io.BytesIO(citibikeZipFile.content))

# And here we are looping through the package and printing the names of the files
for name in zipFiles.namelist():
  print(name)

# Finally, we can take that loaded uncopressed package and use it to just load one file to the Pandas dataframe:
df = pd.read_csv(zipFiles.open('202109-citibike-tripdata.csv'))
```

```python
#Converting the columns to floats
df1['end station latitude'] = df1['end station latitude'].astype(float)
df1['end station longitude'] = df1['end station longitude'].astype(float)
df1['start station latitude'] = df1['start station latitude'].astype(float)
df1['start station longitude'] = df1['start station longitude'].astype(float)
```

```python
df.head(2)
```

```python
#Group by station ID into a new dataframe
start_citibike20 = df1.groupby(['start station name']).agg({'start station latitude':'min', 'start station longitude':'min', 'start station id': 'count'}).reset_index()

start_citibike20.head(5)
```

```python
#Group by station ID into a new dataframe
stop_citibike20 = df1.groupby(['end station name']).agg({'end station latitude': 'min', 'end station longitude': 'min', 'end station id': 'count'}).reset_index()

stop_citibike20.head(5)
```

```python
#Geodataframe for starts 2020
citiStartsGeo20 = gpd.GeoDataFrame(data=start_citibike20, geometry=gpd.points_from_xy(x=start_citibike20['start station longitude'], y=start_citibike20['start station latitude']), crs='epsg:4326')
citiStartsGeo20.head(2)
```

```python
citiStartsGeo20= citiStartsGeo20.rename(columns={'start station id': 'tripCount'})
```

```python
#Geodataframe for stops 2020
citiStopsGeo20 = gpd.GeoDataFrame(data=stop_citibike20, geometry=gpd.points_from_xy(x=stop_citibike20['end station longitude'], y=stop_citibike20['end station latitude']), crs='epsg:4326')
citiStopsGeo20.head(2)
```

```python
citiStopsGeo20 = citiStopsGeo20.rename(columns={'end station id': 'tripCount'})
citiStopsGeo20.head(2)
```

```python
#Group by station ID into a new dataframe
start_citibike21 = df.groupby(['start_station_name']).agg({'start_lat':'min', 'start_lng':'min', 'start_station_id': 'count'}).reset_index()

start_citibike21.head(5)
```

```python
start_citibike21 = start_citibike21.rename(columns={'start_station_id': 'tripCount'})
start_citibike21.head(2)
```

```python
#Geodataframe for starts 2021
citiStartsGeo21 = gpd.GeoDataFrame(data=start_citibike21, geometry=gpd.points_from_xy(x=start_citibike21['start_lng'], y=start_citibike21['start_lat']), crs='epsg:4326')
citiStartsGeo21.head(2)
```

```python
stop_citibike21 = df.groupby(['end_station_name']).agg({'end_lat': 'min', 'end_lng': 'min', 'end_station_id': 'count'}).reset_index()

stop_citibike21.head(5)
```

```python
stop_citibike21 = stop_citibike21.rename(columns={'end_station_id': 'tripCount'})
stop_citibike21.head(2)
```

```python
#Geodataframe for stops 2021
citiStopsGeo21 = gpd.GeoDataFrame(data=stop_citibike21, geometry=gpd.points_from_xy(x=stop_citibike21['end_lng'], y=stop_citibike21['end_lat']), crs='epsg:4326')
citiStopsGeo21.head(2)
```

```python
#expert to file
citiStartsGeo20.to_file('citiStartsGeo20.geojson', driver='GeoJSON')
citiStopsGeo20.to_file('citiStopsGeo20.geojson', driver='GeoJSON')
citiStartsGeo21.to_file('citiStartsGeo21.geojson', driver='GeoJSON')
citiStopsGeo21.to_file('citiStopsGeo21.geojson', driver='GeoJSON')
```
